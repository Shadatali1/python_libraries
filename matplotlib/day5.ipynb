# **Bar Plot (Categorical Data + Comparison)**

```python
import matplotlib.pyplot as plt
```

### What is a Bar Plot?

👉 **Simple definition:**\
A bar plot is used to compare numerical values across different categories

**Key Idea**

👉 One side = categories\
👉 Other side = numbers

| Category | Sales |
| -------- | ----- |
| A        | 100   |
| B        | 150   |
| C        | 120   |

**👉 Question:**

- Which category is highest? ✔️
- Which is lowest? ✔️

👉 Bar plot answers this instantly.

### First Bar Plot

```python
categories = ['A', 'B', 'C']
sales = [100, 150, 120]

fig, ax = plt.subplots()

ax.bar(categories, sales)

ax.set_title("Sales by Category")
ax.set_xlabel("Category")
ax.set_ylabel("Sales")

plt.show()
```

### What Are We Seeing?

👉 Each bar = one category\
👉 Height of bar = value

### Styling Bar

```python
categories = ['A', 'B', 'C']
sales = [100, 150, 120]

fig, ax = plt.subplots()

ax.bar(categories, sales , color = 'blue')

ax.set_title("Sales by Category")
ax.set_xlabel("Category")
ax.set_ylabel("Sales")

plt.show()
```

**Different colors for each bar**

```python
categories = ['A', 'B', 'C']
sales = [100, 150, 120]

fig, ax = plt.subplots()
colors = ['red', 'green', 'blue']
ax.bar(categories, sales , color = colors)

ax.set_title("Sales by Category")
ax.set_xlabel("Category")
ax.set_ylabel("Sales")

plt.show()
```

### Real-World Example
Students Marks by Subject

```python
subjects = ['Math', 'Physics', 'CS']
marks = [85, 78, 92]

fig, ax = plt.subplots()

ax.bar(subjects, marks, color=['blue', 'orange', 'green'])

ax.set_title("Marks by Subject")
ax.set_xlabel("Subjects")
ax.set_ylabel("Marks")

plt.show()
```

### Data Science Thinking

👉 What are we doing?

Comparing categories\
Finding best / worst\
Making decisions

### Example Insight

👉 If CS > Math > Physics\
→ Student is stronger in CS

### Horizontal Bar Plot (IMPORTANT)

```python
subjects = ['Math', 'Physics', 'CS']
marks = [85, 78, 92]

fig, ax = plt.subplots()

ax.barh(subjects, marks, color=['blue', 'orange', 'green'])

ax.set_title("Marks by Subject")
ax.set_xlabel("Subjects")
ax.set_ylabel("Marks")

plt.show()
```

👉 **When to use?**

- Long category names
- Better readability

| Use Case            | Chart     |
| ------------------- | --------- |
| Time trend          | Line plot |
| Category comparison | Bar plot  |


### Mini Statistics Concept

👉 Bar plot shows:
