# **Box Plot (Median + Quartiles + Outliers)**

### What is a Box Plot?

👉 Simple definition:

A box plot shows the summary of data distribution using statistics

**Key Idea**

👉 Instead of showing all values\
👉 It shows important summary points

### Important Statistics (CORE CONCEPT 🔥)

Let’s take data:

```python
marks = [40, 50, 60, 70, 80, 90, 100]
```

**Key Points:**\
🔹 Median (Middle Value)

👉 Middle of data\
40 50 60 [70] 80 90 100

### Quartiles

👉 Divide data into 4 parts:

- Q1 → 25%
- Q2 → Median (50%)
- Q3 → 75%

**Range**

👉 Min → Max

**Outliers (VERY IMPORTANT)**

👉 Values far from normal data

Example:

```python
marks = [40, 50, 60, 70, 80, 90, 200]
# 200 > Outlier
```

### Box Plot Structure
|----|====|----|\
min  Q1  median  Q3  max

👉 Box = Q1 to Q3\
👉 Line inside = Median\
👉 Points outside = Outliers

### First Box Plot

```python
marks = [40, 50, 60, 70, 80, 90, 100]

fig, ax = plt.subplots()

ax.boxplot(marks)

ax.set_title("Marks Distribution")

plt.show()
```

**What Are We Seeing?**

👉 Middle line → median\
👉 Box → spread of data\
👉 Lines → range

**Salary Analysis**

```python
salary = [20000,25000,30000,35000,40000,45000,100000]

fig, ax = plt.subplots()

ax.boxplot(salary)

ax.set_title("Salary Distribution")

plt.show()
```

**Insight**

👉 Most salaries = 20k–45k\
👉 100k = Outlier

| Histogram   | Box Plot      |
| ----------- | ------------- |
| Shows shape | Shows summary |
| Visual      | Analytical    |


### Multiple Box Plots (Comparison)

```python
class_A = [40,50,60,70,80]
class_B = [30,45,55,65,75]

fig, ax = plt.subplots()

ax.boxplot([class_A, class_B])

plt.show()
```

### Homework
✅ **Task 1**

Create:

study_hours = [1,2,2,3,3,4,4,5,6,8,10]

👉 Plot box plot

✅ **Task 2**

Add an outlier:

study_hours.append(15)

👉 Observe change

✅ **Task 3 (Thinking)**

👉 What is median?\
👉 Any outliers?\
👉 Is data balanced?

```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------
# Task 1: Original Data
# -----------------------

study_hours = [1,2,2,3,3,4,4,5,6,8,10]

print("Original Data:", study_hours)

# Calculate median
median_before = np.median(study_hours)
print("Median (before adding outlier):", median_before)

# Plot boxplot (Before outlier)
plt.figure()
plt.boxplot(study_hours)
plt.title("Box Plot - Before Adding Outlier")
plt.ylabel("Study Hours")
plt.grid(True)
plt.show()


# -----------------------
# Task 2: Add Outlier
# -----------------------

study_hours.append(15)

print("\nData After Adding Outlier:", study_hours)

median_after = np.median(study_hours)
print("Median (after adding outlier):", median_after)

# Plot boxplot (After outlier)
plt.figure()
plt.boxplot(study_hours)
plt.title("Box Plot - After Adding Outlier")
plt.ylabel("Study Hours")
plt.grid(True)
plt.show()


# -----------------------
# Task 3: IQR & Outliers
# -----------------------

Q1 = np.percentile(study_hours, 25)
Q3 = np.percentile(study_hours, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("\nQ1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Detect outliers manually
outliers = [x for x in study_hours if x < lower_bound or x > upper_bound]

print("Detected Outliers:", outliers)
```
