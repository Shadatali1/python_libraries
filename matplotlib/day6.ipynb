# **Histogram (Understanding Data Distribution)**

```python
import matplotlib.pyplot as plt
```

### What is a Histogram?

👉 Simple definition:

A histogram shows how data is distributed (spread out)

### Key Idea

👉 It answers:

How many values fall in a range?\
Is data concentrated or spread out?

### Example Thinking
Marks of Students:

```python
marks = [45, 50, 55, 60, 60, 65, 70, 75, 80, 85, 90]
```

👉 **Questions:**

Most students in which range?
Are marks low or high overall?

👉 Histogram answers this.

### First Histogram

```python
marks = [45, 50, 55, 60, 60, 65, 70, 75, 80, 85, 90]

fig, ax = plt.subplots()

ax.hist(marks)

ax.set_title("Marks Distribution")
ax.set_xlabel("Marks")
ax.set_ylabel("Frequency")

plt.show()
```

### What is Happening?

👉 Histogram groups data into bins (ranges)

Example:

40–50\
50–60\
60–70

👉 Then counts how many values fall in each bin

### Bins (VERY IMPORTANT)

👉 Control grouping:

```python
marks = [45, 50, 55, 60, 60, 65, 70, 75, 80, 85, 90]

fig, ax = plt.subplots()

ax.hist(marks , bins = 2)

ax.set_title("Marks Distribution")
ax.set_xlabel("Marks")
ax.set_ylabel("Frequency")

plt.show()
```

**Understanding Bins**
- Small bins → more detail
- Large bins → more general view

### Real-World Example
Daily Income Data

```python
income = [200,220,250,270,300,320,350,370,400,450,500]

fig, ax = plt.subplots()

ax.hist(income, bins=4)

ax.set_title("Income Distribution")
ax.set_xlabel("Income")
ax.set_ylabel("Frequency")

plt.show()
```

### Data Science Thinking

👉 Histogram helps you see:

- Data concentration
- Spread
- Skewness (advanced concept later)

### Mini Statistics Concepts 🔥
📌 Frequency

👉 Number of values in each bin

### Distribution

👉 Shape of data:

Common types:\
✅ Normal (balanced)\
   ▂▃▅▇▅▃▂\
❌ Skewed (one side heavy)\
▇▆▅▃▂

```python
income = [200,220,250,270,300,320,350,370,400,450,500]

fig, ax = plt.subplots()

ax.hist(income, bins=4 , color = 'blue')

ax.set_title("Income Distribution")
ax.set_xlabel("Income")
ax.set_ylabel("Frequency")
ax.grid(True)

plt.show()
```

### When to Use Histogram?

👉 Use when:

- One numerical variable
- Want to see distribution

### When NOT to use
- Categories → use bar chart
- Relationship → use scatter plot
